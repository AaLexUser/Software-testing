```python
def sec(x):
    cosRes = math.sin(x)
    if abs(cosRes) < 1e-5:
        return math.nan
    return 1/math.cos(x)
def csc(x):
    sinRes = math.sin(x)
    if abs(sinRes) < 1e-5:
        return math.nan
    return 1/math.sin(x)

def log3(x):
    return math.log(x, 3)
def log5(x):
    return math.log(x, 5)

x_list = np.arange(-2*math.pi, 2*math.pi, math.pi/100)
x_list = np.append(x, [np.nan])
y_list = np.zeros(len(x_list))
# complex function
for x, y in zip(x_list, y_list):
    if(x <= 0):
        y = sec(x) - csc(x)
    else:
        y = log3(x) - log5(x) - log5(x)

df = pd.DataFrame({'x': x_list, 'y': y_list})
df.to_csv('../src/test/resources/integration/myFunc.csv', index=False, na_rep='NaN')
display(df)

plt.plot(x_list, y_list)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
#minor grid
plt.minorticks_on()
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```
